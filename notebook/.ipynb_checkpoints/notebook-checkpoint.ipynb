**Removing the outliers**

```python
#functions for removing outliers
def remove_outliers(data,labels):
    for label in labels:
        q1 = data[label].quantile(0.25)
        q3 = data[label].quantile(0.75)
        iqr = q3 - q1
        upper_bound = q3 + 1.5 * iqr
        lower_bound = q1 - 1.5 * iqr
        data[label] = data[label].mask(data[label]< lower_bound, data[label].median(),axis=0)
        data[label] = data[label].mask(data[label]> upper_bound, data[label].median(),axis=0)

    return data
```

```python
df = remove_outliers(df, numerical)
```

```python
df = df.drop(columns=['CustomerID'])
df.columns
```

**Encoding**

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
for cat in categorical:
    df[cat] = label_encoder.fit_transform(df[cat])
df.head()
```

**Train Test**

```python
from sklearn.model_selection import train_test_split

X = df.drop('Churn',axis=1).values
y = df.Churn.values
# spliting the data into test and train
x_train, x_test , y_train, y_test = train_test_split(X, y , test_size=0.3, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

**Scaling**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
X_train.shape
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
```

```python
svc = SVC(C=1.0, kernel='rbf', degree=3, gamma='scale')
# svc = SVC(kernel='rbf', decision_function_shape='ovr')
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
print(f'SVC-Accuracy: {accuracy_score(y_test, y_pred)}')
print('Classification report: ', classification_report(y_test, y_pred))
print('{}'.format(classification_report(y_test, y_pred)))
print('Confusion Matrix \n ', confusion_matrix(y_test, y_pred))
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
y_pred = tree.predict(X_test)
print(f'Decision Tree-Accuracy: {accuracy_score(y_test, y_pred)}')
print('Classification report: ', classification_report(y_test, y_pred))
print('{}'.format(classification_report(y_test, y_pred)))
print('Confusion Matrix \n ', confusion_matrix(y_test, y_pred))
```

**Hypertuning GridSearchCV**

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Define the parameter grid for GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 150],  
    'learning_rate': [0.01, 0.1, 0.3], 
    'max_depth': [3, 6, 9], 
    'min_child_weight': [1, 3, 5], 
    # 'subsample': [0.8, 0.9, 1],  
    'colsample_bytree': [0.8, 0.9, 1], 
    'gamma': [0, 0.1, 0.3],  
    # 'alpha': [0, 0.1, 0.5],  
    'lambda': [1, 1.5, 2], 
}

# Define the XGBClassifier model
xgb_model = XGBClassifier(
    # n_jobs=4,
    scale_pos_weight=1,
    seed=27
)

# Initialize GridSearchCV
grid = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid, 
    scoring='roc_auc',
    # cv=3,
    n_jobs=-1,
    verbose=1
)

# Fit GridSearchCV
grid.fit(X_train, y_train)

# Display the best parameters and best score
print('Best Grid Search Parameters :', grid.best_params_)
print('Best Grid Search Score :', grid.best_score_)

# Predict using the best model found by GridSearchCV
y_pred = grid.best_estimator_.predict(X_test)

# Calculate accuracy and display the results
accuracy = accuracy_score(y_test, y_pred)
print(f'Grid-Accuracy: {accuracy * 100:.2f}%')
print('Classification report: ', classification_report(y_test, y_pred))
print('Confusion Matrix \n', confusion_matrix(y_test, y_pred))
```

**Deep Learning**

```python
# !pip install keras
# !pip install tensorflow
```

```python
import tensorflow as tf
from sklearn.metrics import classification_report, confusion_matrix

# Define the model using tf.keras (which is Keras within TensorFlow)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, input_dim=X_train.shape[1], activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=64)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Get predictions and convert them to binary labels (0 or 1)
y_pred = (model.predict(X_test) > 0.5).astype(int)

# Print classification report and confusion matrix
print('Classification report: ', classification_report(y_test, y_pred))
print('Confusion Matrix \n', confusion_matrix(y_test, y_pred))
```
